```python
#Rename & Sneak Case
cols = []
for i in range(len(data.columns)):
    cols.append(data.columns[i].lower().replace(' ', '_'))
data.columns = cols
cols
```

```python
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'])
data['effective_to_date'] = pd.to_datetime(data['effective_to_date'], format='%Y-%m-%d')
data['effective_to_date']
```

```python
categorical_df = data.select_dtypes(object)
categorical_df.head()
```

```python
# Education
categorical_df['education'] = categorical_df['education'].fillna('Above High School')
categorical_df['education'] = np.where(categorical_df['education'].isin(['Bachelor', 'College', 'Master', 'Doctor']) ,'Above High School', categorical_df['education'])
categorical_df['education'].value_counts()
```

```python
categorical_df['location_code'] = categorical_df['location_code'].fillna('Suburban')
categorical_df['location_code'] = np.where(categorical_df['location_code'].isin(['Urban']) ,'Suburban', categorical_df['location_code'])
categorical_df['location_code'].value_counts()
```

```python
categorical_df['sales_channel'].value_counts()
```

```python
# luxury or Sport category
categorical_df['vehicle_class'] = categorical_df['vehicle_class'].fillna('Luxury or Sport')
categorical_df['vehicle_class'] = np.where(categorical_df['vehicle_class'].isin(['Luxury Car', 'Luxury SUV', 'Sports Car', 'SUV']) ,'Luxury or Sport', categorical_df['vehicle_class'])
categorical_df['vehicle_class'].value_counts()
```

```python
categorical_df['marital_status'].value_counts()
```

```python
categorical_df['policy'].value_counts()
```

```python
categorical_df['policy'] = categorical_df['policy'].fillna('Personal')
categorical_df['policy'] = np.where(categorical_df['policy'].isin(['Personal L3', 'Personal L2', 'Personal L1']) ,'Personal', categorical_df['policy'])
```

```python
categorical_df['policy'] = categorical_df['policy'].fillna('Corporate')
categorical_df['policy'] = np.where(categorical_df['policy'].isin(['Corporate L3', 'Corporate L2', 'Corporate L1']) ,'Corporate', categorical_df['policy'])
```

```python
categorical_df['policy'] = categorical_df['policy'].fillna('Special')
categorical_df['policy'] = np.where(categorical_df['policy'].isin(['Special L1', 'Special L2', 'Special L3']) ,'Special', categorical_df['policy'])
categorical_df['policy'].value_counts()
```
